```python
# now handle the missing values for both the categorical and numerical features
# first check how many nan values for categorical features

categorical_nan = []
for feature in data.columns:
    if data[feature].isnull().sum() > 1 and data[feature].dtypes=='O': 
        categorical_nan.append(feature)
```

```python
for feature in categorical_nan:
    print(feature, np.round(data[feature].isnull().mean(),4),'% missing values')
```

```python
# now in the case of categorical features, replace the missing records with missing label
def replace_cat_features(data,categorical_nan):
    data_copy = data.copy()
    data[categorical_nan]=data[categorical_nan].fillna('Missing')
    return data

data_copy = replace_cat_features(data,categorical_nan)
```

```python
numerical_nan = []
for feature in data.columns:
    if data[feature].isnull().sum() > 1 and data[feature].dtypes!='O': 
        numerical_nan.append(feature)
```

```python
for feature in numerical_nan:
    print(feature, np.round(data[feature].isnull().mean(),4),'% missing values')
```

```python
# in case of outliers with missing values features , we replace the missing values with median or mode
for feature in numerical_nan:
    median_feature = data[feature].median()
    data[feature+'id'] = np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_feature,inplace=True)
    
```

```python
# handle year part
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    data[feature] = data['YrSold']- data[feature]
```

```python
import numpy as np
num_features_outliers =['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice'] 
for feature in num_features_outliers:
    data[feature] = np.log(data[feature])
```

```python
data.head()
```

```python
# Remove categorical features that has less than 1% impact on the target variable.
categorical_features = []
for feature in data.columns:
    if data[feature].dtype=='O':
        categorical_features.append(feature)
        
```

```python
for feature in categorical_features:
           temp= data.groupby(feature)['SalePrice'].count()/len(data)
           temp_df = temp[temp>0.01].index
           data[feature] = np.where(data[feature].isin(temp_df),data[feature],'Rare_val') 
       
    
```

```python
categorical_features
```

```python
for feature in categorical_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
# feature scaling
scale_feature = [feature for feature in data if feature not in ['Id','SalesPrice']]
data[scale_feature]    
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[scale_feature])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(data[scale_feature]), columns=scale_feature)],
                    axis=1)
```
